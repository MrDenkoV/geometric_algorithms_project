# QuadTree

```python
import quadtree
```

## Create points

```python
points_set = quadtree.get_points(0)
```

## QuadTree implementation

```python
tree = quadtree.QuadTree(points_set)

points = list(map(quadtree.Point.get_tuple, points_set))
solution = tree.find()
print(solution)
```

```python
scenes = [Scene([PointsCollection(points),
                     PointsCollection(solution, color='red', marker="x")])]
plot = Plot(scenes)
plot.draw()
```
